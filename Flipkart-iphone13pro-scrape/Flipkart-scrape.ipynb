```python
import requests
```

```python
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
source = requests.get('https://www.flipkart.com/search?q=iphone%2013%20pro&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off').content

soup = BeautifulSoup(source, 'lxml')

csv_file = open('iphone-13pro.csv','w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title','Price','Rating'])

for match,match1,match2 in zip(soup.find_all('div', class_='_4rR01T'),
                               soup.find_all('div', class_='_3LWZlK'),
                               soup.find_all('div', class_='_30jeq3 _1_WHN1')):
                              
    print(f' Title: {match.text}\n Rating: {match1.text}\n Price: {match2.text}')
    print()
    
    csv_writer.writerow([match.text, match2.text, match1.text])

csv_file.close()
print("Success!! CSV file created")

# match1 = soup.find('div', class_='_3LWZlK')
# print(f'Rating: {match1.text}')

# match2 = soup.find('div', class_='_30jeq3 _1_WHN1')
# print(f'Price: {match2.text}')
```
